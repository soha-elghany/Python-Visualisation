# Visualizing 40 Years of Music Industry Sales

```python
import numpy as np
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("/Users/soha/Documents/Python/Python Visualisations/Music Industry/MusicData.csv", parse_dates=["Year"], thousands=',')

df
```

```python
df.dropna(subset=['Value (Actual)'],inplace = True)
df
```

```python
df.astype({"Value (Actual)": "float"})
df["Value (Actual)"].dtype
```

```python
newdf = df[(df.Metric == "Value (Adjusted)")]
newdf
```

```python
def f(row):
    if (
        row['Format'] =='Cassette'
        or row['Format'] =='Cassette Single'
        or row['Format'] =='SACD'
    ):
        val = 'Cassette'
    
    elif (
        row['Format'] =='CD'
        or row['Format'] =='CD Single'
    ):
        val = 'CD'
    
    elif (
        row['Format'] =='Download Album'
        or row['Format'] =='Download Music Video'
        or row['Format'] =='Download Single'
        or row['Format'] =='Limited Tier Paid Subscription'
        or row['Format'] =='On-Demand Streaming (Ad-Supported)'
        or row['Format'] =='Other Ad-Supported Streaming'
        or row['Format'] =='Other Digital'
        or row['Format'] =='Paid Subscription'
        or row['Format'] =='Paid Subscriptions'
        or row['Format'] =='Ringtones & Ringbacks'
        or row['Format'] =='SoundExchange Distributions'
        or row['Format'] =='Synchronization'
    ):
        val = 'Download'
    elif (
        row['Format'] =='Vinyl Single'
        or row['Format'] =='LP/EP'
    ):
        val = 'Vinyl'
    else:
        val = 'Others'
    return val
```

```python
newdf['Group'] = newdf.apply(f, axis = 1)
```

```python
area = alt.Chart(newdf).mark_area(opacity = 0.7).configure_axis(grid=False).encode(
    x="Year:T",
    y=alt.Y("sum(Value (Actual)):Q", stack = None),
    color=alt.Color(
        "Group:N",
        scale=alt.Scale(
            domain=[
                "CD",
                "Cassette",
                "Download",
                "Others",
                "Vinyl",
            ],
            range=["#FF3366", "#000E48", "#663BC9", "#DED4DA", "#4CC9f0"],
        ),
    ),  
    tooltip=["Group", "sum(Value (Actual)):Q", "Year:T"]
).properties(
    title='40 Years: Music Industry'
)

area
```

```python
area.save("music_industry.html")
```
